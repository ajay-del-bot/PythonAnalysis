```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("appstore_games.csv")
df.head(10)
```

```python
len(df)
```

<h2>Cleaning of data</h2>

```python
df.dropna(subset=['Average User Rating', 'User Rating Count'], inplace=True)
df.drop(columns=['URL', 'Icon URL'], inplace=True)
```

```python
len(df)
```

```python
sorted_df = df.sort_values(by=['Average User Rating', 'User Rating Count'], ascending=False)
```

```python
sorted_df.head(10)
```

```python
sorted_df.reset_index(inplace = True) 
```

```python
sorted_df.head()
```

```python
df2 = sorted_df[['Name','Average User Rating', 'User Rating Count', 'Price', 'Genres']]
```
